```python
import torch
from torchvision import datasets, transforms
```

```python
celeb = datasets.CelebA(root='./data', split='train', download=True)
```

```python
figure = plt.figure(figsize=(8, 8))
cols, rows = 3, 3
for i in range(1, cols * rows + 1):
    sample_idx = torch.randint(len(celeb), size=(1,)).item()
    img, label = celeb[sample_idx]
    figure.add_subplot(rows, cols, i)
    plt.axis("off")
    plt.imshow(img, cmap="gray")
plt.show()
```
